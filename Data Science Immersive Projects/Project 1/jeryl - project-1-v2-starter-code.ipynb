Your function will:

5. Return the player's power value.

````python
def play_pow(dict1, dict2, p_id):
    # print poke levels
    for poke in dict1[p_id]['player_pokemon'].keys():
        print str(dict1[p_id]['player_name']) + "'s Pokemon: " + str(dict1[p_id]['player_pokemon'][poke]['name']) + " is level ?."
    
    # Aggregate attack and defense values, sp att, sp def and speed : this will be the base stats and player power value as well
    agg_att = []
    for key in dict1[p_id]['player_pokemon'].keys():
        pokestats = dict2[key]['attack'] + dict2[key]['defense'] + dict2[key]['special_attack'] + dict2[key]['special_defense'] + dict2[key]['speed']
        agg_att.append(pokestats)
        
    play_pow_sum = sum(agg_att)
    print str(dict1[p_id]['player_name']) + "'s power is " + str(play_pow_sum) + "."
    return play_pow_sum

print play_pow(players, pokedex, 2)
print play_pow(players, pokedex, 12345)
````

### 9.1

What is the population mean and standard deviation of the "Total" attribute for all characters in the Pokedex?



````python
# calculate popn mean and s.d. Sum totals and total - mean

grand_tot = []
grand_xbar = []

for key in comp_poke.keys():
    grand_tot.append(comp_poke[key]['Total'])
    
mean = sum(grand_tot) / len(grand_tot)

for key in comp_poke.keys():
    grand_xbar.append((comp_poke[key]['Total'] - mean)**2)
#print grand_xbar
    
sd = (sum(grand_xbar) / (len(grand_tot)))**0.5

print "Population mean is: " + str(mean)
print "Population S.D. is " + str(sd)
````
