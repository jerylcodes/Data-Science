## my Grocery Finder Prototype

This is a demo notebook to test classifying images based on a Convolutional Neural Network.

```python
import os, cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import numpy as np
```

```python
from keras import backend as K
K.set_image_dim_ordering('tf')

import h5py

from keras.utils import np_utils
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from keras.optimizers import SGD,RMSprop,adam
from keras.preprocessing import image

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
# Loading model weights for best epoch , epoch = 202

from keras.models import load_model
loaded_model = load_model('./dropall weights/Best-weights-my_model-202-0.0525-0.9820.hdf5')
```

```python
def mygroceryfinder(imgfile):
    legend = {0:'Wigglys',
        1:'Big red',
        2:'Hersheys chocolate',
        3:'Twix',
        4:'Snickers',
        5:'Tic tac',
        6:'Dentyne ice',
        7:'Starburst',
        8:'Rockstar energy drink',
        9:'Crunch',
        10:'Lays chips',
        11:'Milano biscuits',
        12:'Starbucks frappucino',
        13:'Obertos beef jerky',
        14:'Honey nut cheerios',
        15:'Tostitos scoops',
        16:'Ruffles'}
    input_img=cv2.imread(imgfile)
    input_img_resize=cv2.resize(input_img,(128,128))
    list1 = []
    list1.append(input_img_resize)
    img_data = np.array(list1)
    img_data = img_data.astype('float32')
    img_data /= 255
    grocery = loaded_model.predict_classes(img_data)
  
    image = mpimg.imread(imgfile)
    plt.imshow(image)
    plt.show()
    print('Your input image is: ', legend[grocery[0]])
```

```python
mygroceryfinder('./datasets/conf/13/web1.jpg')
```
